```python
def get_dates(x):
    try:
        return re.search(r'^https:\/\/techcrunch\.com\/(\d{4}\/\d{2}\/\d{2})', x).group(1)
    except:
        return 0
```

```python
# new column "publiched_at"
data["published_at"] = data.link.apply(get_dates)
```

```python
data[data.published_at == 0]
```

```python
data = data[data.published_at != 0]
```

```python
# get funding round
def get_funding_round(row):
    try:
        return re.search(r'[Ss]eries.[ABCDabcd]', row["title"]).group(0)
        
    except:
        try:
            return re.search(r'[Ss]eries.[ABCDabcd]', row["excerpt"]).group(0)
            print ("test")
        except:
            return 0
        
data["funding_round"] = data.apply(lambda row: get_funding_round(row), axis = 1)
```

```python
data[data["funding_round"] == 0]
```

```python
def get_money_raised(row):
    try:
        return re.search(r'\D\d{1,}\.*\d*\s*[Mm][illion]*', row['title']).group(0)
    except:
        try:
            return re.search(r'\D\d{1,}\.*\d*\s*[Mm][illion]*', row['excerpt']).group(0)
        except:
            return 0

data["money_raised"] = data.apply(lambda row: get_money_raised(row), axis = 1)
```

```python
data[data["money_raised"] == 0]
```

```python
# drop money_raised == 0 AND funding_round == 0
data.drop(data[ (data.funding_round == 0) & (data.funding_round == 0)].index, inplace = True)
```
